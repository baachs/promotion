```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# path muss der Pfad zu deinen Daten von dem Verzeichnis dieser Datei aus sein
path = "DatenSchaabBabette-3.xlsx"
df = pd.read_excel(path)
df.shape 
```

```python
# path muss der Pfad zu deinen Daten von dem Verzeichnis dieser Datei aus sein
path = "DatenSchaabBabette-5.xlsx"
df = pd.read_excel(path)
df.shape 
```

```python
# 500ms pre-trigger Zeit abziehen
# ACHTUNG: Diese Zelle nur 1x ausf√ºhren
cols = df.columns.tolist()
lat_variables = [col for col in cols if "_Lat_" in col] # das sind die Latenz Spalten
for col in lat_variables:
    df[col] -= 500
print("DONE!")
```

```python
# ACHTUNG: HIER SOLLTE ~286 STEHEN, FALLS NICHT, WURDE DIE PRE-TRIGGER ZEIT FALSCH ABGEZOGEN
df[lat_variables].mean().mean()
```

```python
patienten = df[df['Status '] == 1]
patienten.shape
```

```python
gesunde = df[df['Status '] == 0]
gesunde.shape
```

# Boxplot zu Latenzen Patienten vs. Gesunde

```python
variable = "Suess_Cz_Lat_P1"
array_a = patienten[variable].dropna()
array_b = gesunde[variable].dropna()

 
fig = plt.figure(figsize =(10, 7))
 
plt.boxplot([array_a, array_b])
# plt.title(f"Boxplot f√ºr {variable}")
plt.title("S√º√ü")
plt.ylim(550, 950)
plt.ylabel('Latenz [ms]')
plt.xticks([1, 2], ['Patienten', 'Gesunde'])
 
# show plot
plt.show()
```

```python
variable = "Salz_Cz_Lat_P1"
array_a = patienten[variable].dropna()
array_b = gesunde[variable].dropna()

 
fig = plt.figure(figsize =(10, 7))
 
plt.boxplot([array_a, array_b])
# plt.title(f"Boxplot f√ºr {variable}")
plt.title("Salzig")
plt.ylim(550, 950)
plt.ylabel('Latenz [ms]')
plt.xticks([1, 2], ['Patienten', 'Gesunde'])
 
# show plot
plt.show()
```

# Boxplots zu Zeitdifferenzen zwischen GEPs

```python
# Berechne die Zeitdifferenzen
df['Zeitdifferenz_Suess_Fz_P1_nach_N1'] = df['Suess_Fz_Lat_N1'] - df['Suess_Fz_Lat_P1']
df['Zeitdifferenz_Suess_Fz_N1_nach_P2'] = df['Suess_Fz_Lat_P2'] - df['Suess_Fz_Lat_N1']
df['Zeitdifferenz_Suess_Cz_P1_nach_N1'] = df['Suess_Cz_Lat_N1'] - df['Suess_Cz_Lat_P1']
df['Zeitdifferenz_Suess_Cz_N1_nach_P2'] = df['Suess_Cz_Lat_p2'] - df['Suess_Cz_Lat_N1']
df['Zeitdifferenz_Suess_Pz_P1_nach_N1'] = df['Suess_Pz_Lat_N1'] - df['Suess_Pz_Lat_P1']
df['Zeitdifferenz_Suess_Pz_N1_nach_P2'] = df['Suess_Pz_Lat_P2'] - df['Suess_Pz_Lat_N1']

df['Zeitdifferenz_Salz_Fz_P1_nach_N1'] = df['Salz_Fz_Lat_N1'] - df['Salz_Fz_Lat_P1']
df['Zeitdifferenz_Salz_Fz_N1_nach_P2'] = df['Salz_Fz_Lat_P2'] - df['Salz_Fz_Lat_N1']
df['Zeitdifferenz_Salz_Cz_P1_nach_N1'] = df['Salz_Cz_Lat_N1'] - df['Salz_Cz_Lat_P1']
df['Zeitdifferenz_Salz_Cz_N1_nach_P2'] = df['Salz_Cz_Lat_p2'] - df['Salz_Cz_Lat_N1']
df['Zeitdifferenz_Salz_Pz_P1_nach_N1'] = df['Salz_Pz_Lat_N1'] - df['Salz_Pz_Lat_P1']
df['Zeitdifferenz_Salz_Pz_N1_nach_P2'] = df['Salz_Pz_Lat_P2'] - df['Salz_Pz_Lat_N1']
```

```python
# Jetzt schauen wir uns unterschiede zwischen patienten/gesunden an
gesunde = df[df['Status '] == 0]
patienten = df[df['Status '] == 1]
print("Patienten Dataframe:", patienten.shape)
print("Gesunden Dataframe:", gesunde.shape)
```
